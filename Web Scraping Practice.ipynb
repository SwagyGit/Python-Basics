# Web Scraping for multiple pages

```python
import requests
import bs4
```

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'

for i in range(1,11):
    target_url = base_url.format(i)
    
    page = requests.get(target_url)
    page_content = bs4.BeautifulSoup(page.text,'lxml')
    books = page_content.select('.product_pod')
    for book in books:
        
        if len(book.select('.star-rating.Two'))!=0:
            print(book.select('a')[1]['title'])
        
```

```python
page = requests.get('http://books.toscrape.com/catalogue/page-3.html')
page_content = bs4.BeautifulSoup(page.text,'lxml')
books = page_content.select('.product_pod')
book = page_content.select('.product_pod')[0]
book.select('.star-rating.Two')
```
