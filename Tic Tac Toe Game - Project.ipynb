```python
# Welcome to Tic Tac Toe Game
def start_game():
    
    decision = False
    # Display function
    board = game_display()
    
    #Toss for Player Selection
    toss_a_coin()

    # Ask for input string
    input1,input2 = player_input()
    marker_dic = {'Player1_marker':input1,'Player2_marker':input2}
    
   # print('After player_input')

    # Ask for Position
    p1,p2 = position_in_board(board)
    position_dic = {'Player1_position':p1,'Player2_position':p2}

   # print('After position_in_boad')
    
    # Updating Board
   # print('calling updated_game_display 1st')
    updated_board = updated_game_display(board,marker_dic,position_dic)
    #print('after updated_game_display 1st: {}'.format(updated_board))
    
    
    # Win or Tie
    while decision == False:
        winner,decision = decision_time(updated_board,marker_dic)
       ## print('after decision_time decision is: {}'.format(decision))
        
        if decision:
            if winner == 'Tie':
                print("It's a tie!")
            else:
                print(f'The winner of the game: {winner}')
            #break
            
        else:
            p3,p4 = position_in_board(updated_board)
            position_dic = {'Player1_position':p3,'Player2_position':p4}
            updated_board = updated_game_display(updated_board,marker_dic,position_dic)
            decision = False
    
    #Ask if want to continue the game
    exit_game()
```

```python
#Toss for Player Selection
import random
import time
def toss_a_coin():
    
    ask_flag = True
    toss = ('Heads','Tails')
    
    print("Let's toss the coin and see who is the player1")
    print('Please select heads or tails')
    
    
    while ask_flag:
        
        ask = input('Are you ready [Yes/No]: ')
        
        if ask.upper() == 'YES':
            toss_decision = random.choice(toss)
            time.sleep(1)
            print('And the result is........')
            time.sleep(1)
            print(toss_decision.upper())
            ask_flag = False
            time.sleep(3)
        
        elif ask.isdigit():
            print('Please enter Yes or No')
            continue
        
        elif ask.upper() == 'NO':
            ask_flag = False
            exit_game()
            break
        
        else:
            print('Please enter Yes or No')
            continue
```

```python
# Display function

def game_display():
    
    game_input_list = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    print('The Board::')
    
    print('   |     |   ')
    print(game_input_list[0]+ '  |  ' +game_input_list[1]+ '  |  ' + game_input_list[2])
    print('   |     |   ')
    print('-------------')
    print('   |     |   ')
    print(game_input_list[3]+ '  |  ' +game_input_list[4]+ '  |  ' + game_input_list[5])
    print('   |     |   ')
    print('-------------')
    print('   |     |   ')
    print(game_input_list[6]+ '  |  ' +game_input_list[7]+ '  |  ' + game_input_list[8])
    print('   |     |   ')
    
    return game_input_list    
```

```python
#Updating Board
from IPython.display import clear_output

def updated_game_display(updated_game_input_list,marker_dic,position_dic):
    
    position1 = position_dic['Player1_position']
    if position_dic['Player2_position'] != 10:
        position2 = position_dic['Player2_position']
    
   # print('In updated_game_display')
    
    #Updating Player1 position
    if updated_game_input_list[position1] == ' ':
        updated_game_input_list[position1] = marker_dic['Player1_marker']
    if position_dic['Player2_position'] != 10:
        if updated_game_input_list[position2] == ' ':
            updated_game_input_list[position2] = marker_dic['Player2_marker']
        
    
    #clear_output()
    print('\nThe Updated Board:')
    print('==================')
    print('   |     |   ')
    print(updated_game_input_list[0]+ '  |  ' +updated_game_input_list[1]+ '  |  ' +updated_game_input_list[2])
    print('   |     |   ')
    print('-------------')
    print('   |     |   ')
    print(updated_game_input_list[3]+ '  |  ' +updated_game_input_list[4]+ '  |  ' +updated_game_input_list[5])
    print('   |     |   ')
    print('-------------')
    print('   |     |   ')
    print(updated_game_input_list[6]+ '  |  ' +updated_game_input_list[7]+ '  |  ' +updated_game_input_list[8])
    print('   |     |   ')
    
    return updated_game_input_list
    
```

```python
# Win or Tie
from IPython.display import clear_output

def decision_time(updated_board,marker_dic):
    
    marker1 = marker_dic['Player1_marker']
    marker2 = marker_dic['Player2_marker']
    game_decision = False
    new_winner = ''
    
   # print('In decision_time: {}'.format(updated_board))
    
    #Check for winner
    if updated_board.count(' ')==0:
        #print('If there is no emptyspace')
        #Decision Cases
        if updated_board[0] == updated_board[1] == updated_board[2]:
            new_winner = updated_board[0]
        elif updated_board[3] == updated_board[4] == updated_board[5]:
            new_winner = updated_board[3]
        elif updated_board[6] == updated_board[7] == updated_board[8]:
            new_winner = updated_board[6]
        elif updated_board[0] == updated_board[4] == updated_board[8]:
            new_winner = updated_board[4]
        elif updated_board[2] == updated_board[4] == updated_board[6]:
            new_winner = updated_board[4]
        elif updated_board[0] == updated_board[3] == updated_board[6]:
            new_winner = updated_board[0]
        elif updated_board[1] == updated_board[4] == updated_board[7]:
            new_winner = updated_board[1]
        elif updated_board[2] == updated_board[5] == updated_board[8]:
            new_winner = updated_board[2]
        else:
            new_winner = 'Tie'
            game_decision = True
            return new_winner,game_decision
            
    elif updated_board.count(' ')<=3:
       # print('If there is emptyspace')
        #Decision Cases
        if updated_board[0] == updated_board[1] == updated_board[2] and updated_board[0] != ' ':
           # print('1')
            new_winner = updated_board[0]
        elif updated_board[3] == updated_board[4] == updated_board[5]  and updated_board[3] != ' ':
           # print('2')
            new_winner = updated_board[3]
        elif updated_board[6] == updated_board[7] == updated_board[8]  and updated_board[6] != ' ':
           # print('3')
            new_winner = updated_board[6]
        elif updated_board[0] == updated_board[4] == updated_board[8]  and updated_board[0] != ' ':
          #  print('4')
            new_winner = updated_board[4]
        elif updated_board[2] == updated_board[4] == updated_board[6]  and updated_board[2] != ' ':
          #  print('5')
            new_winner = updated_board[4]
        elif updated_board[0] == updated_board[3] == updated_board[6]  and updated_board[0] != ' ':
           # print('6')
            new_winner = updated_board[0]
        elif updated_board[1] == updated_board[4] == updated_board[7]  and updated_board[1] != ' ':
           # print('7')
            new_winner = updated_board[1]
        elif updated_board[2] == updated_board[5] == updated_board[8]  and updated_board[2] != ' ':
           # print('8')
            new_winner = updated_board[2]
        else:
            return new_winner,game_decision
    
    else:
        return new_winner,game_decision
    
    #print('after decision is true new_winner: {} and marker: {}'.format(new_winner,marker1))
    if new_winner == marker1:
        new_winner = 'Player1'
        game_decision = True
    else:
        new_winner = 'Player2'
        game_decision = True
        
        
    return new_winner,game_decision
```

```python
#Assigning Players with their symbols

def player_input():
    
    #Declairing variables
    player2_input = ''
    acceptable_choice = ['X','O']
    choice = False
        
    while choice == False:
        #Accepting input
        player1_input = input("Player1: Select your choice of marker: ['X' or 'O'] ")
        player1_input = player1_input.upper()
        #Check for String or Digit
        if player1_input.isdigit():
            print('Sorry, it is a Digit! Please enter an acceptable value [X or O]')
            continue
       
        #Check for correct string        
        if player1_input not in acceptable_choice:
            print('Sorry, unacceptable! Please enter an acceptable character [X or O]')
    
        #Check for correct string    
        elif player1_input in acceptable_choice:
            
            choice = True
            #Assigning player2
            if player1_input == 'X':
                player2_input = 'O'
            else:
                player2_input = 'X'
        

         
    print(f'Player1 your symbol is: {player1_input}')
    print(f'Player2 your symbol is: {player2_input}')
    
    return player1_input,player2_input
```

```python
# Ask for Position
def position_in_board(game_board):
    
    checker1 = False
    checker2 = False
        
    while checker1 == False:
        
        position1 = input('Player1: Select your position: [1-9] ')
        checker1 = position_checker(position1,game_board)
        
        if checker1:
            position1 = int(position1)-1
            break
        else:
            continue
    
    while checker2 == False:
        
        if game_board.count(' ')==1:
            return position1,10
        else:
            position2 = input('Player2: Select your position: [1-9] ')
            checker2 = position_checker(position2,game_board)
        
            if checker2:
                position2 = int(position2)-1
                if position1 == position2:
                    print('You must select a different position!')
                    checker2=False
                    continue
                else:
                    break
            else:
                continue
   # print(f'Player1 your position is: {position1}')
   # print(f'Player2 your symbol is: {position2}')
    return position1,position2
    
```

```python
def position_checker(position,board):
    
    if position.isdigit():
        position = int(position)-1
        
        if position in range(0,9):
            #Check for empty position
            if board[position] == ' ':
                return True            
            else:
                print('You must select an empty position!')
                return False
        else:
            print('Sorry, not in the acceptable range. Please select your position 1-9')
            return False
    else:
        print('Sorry, it is not a digit. Please select your position 1-9')
        return False
```

```python
#Exit Decision
from IPython.display import clear_output
def exit_game():
    
    check_exit_flag = False
    
    while not check_exit_flag:
        exit_decision = input('\nDo you want to play again? [Y/N]: ')
        if exit_decision.isdigit():
            print('Sorry, digit is not acceptable! Please enter Y or N')
        
        elif exit_decision.upper() not in ['Y','N']:
            print('Sorry, Please enter Y or N')
            
        else:
            check_exit_flag = True
    
    if exit_decision.upper() == 'Y':
        clear_output()
        print("Let's play the game!")
        start_game()
    else:
       # clear_output()
        print('GAME OVER.\nWell played!')
```
